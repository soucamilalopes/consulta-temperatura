```python
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Leitura inicial para verificar os dados disponíveis no arquivo
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------

#arquivo = open("Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv","r")
#for linha in arquivo:
 # print(linha)
#arquivo.close()



#------------------------------------------------------------------ PARTE 1 - TRATAMENTO DO ARQUIVO ----------------------------------------------------------------

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para leitura do arquivo e retirada do \n
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def CargaDados(nome):
  arquivo = open(nome,"r")
  dados = [] #criação de uma lista vazia para armazenar os dados
  for linha in arquivo:
    linha1 = linha[:-1] #retirada do \n em cada linha do arquivo
    dados.append(linha1)
  arquivo.close()
  return dados



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para escrever os dados da lista no console e verificar os dados
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def escreveLista(lista):
  for item in lista:
    print(item) #impressão da lista no console



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para transformar e estruturar os dados em tuplas
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def transformaDados(linha):
  itens = linha.split(",") #determinando a vírgula como separador
  dia, mes, ano = itens[0].split("/") #determinano a / como separador da data, pois será necessário acessar separadamente o mês e ano depois
  ano = int(ano) #dados de ano transformados em inteiro e incluídos na lista
  mes = int(mes) #dados de mes transformados em inteiro e incluídos na lista
  itensTransformados = [dia, mes, ano] #dados de dia, mês e ano incluídos
  itensTransformados.append(float(itens[1])) #dados de precipitação transformados em float e incluídos na lista
  itensTransformados.append(float(itens[2])) #dados de temp_max transformados em float incluídos na lista
  itensTransformados.append(float(itens[3])) #dados de temp_min transformados em float incluídos na lista
  itensTransformados.append(float(itens[6])) #dados de umidade_relativa_do_ar transformados em float incluídos na lista
  itensTransformados.append(float(itens[7])) #dados de velocidade_do_vento transformados em float incluídos na lista
  return itensTransformados



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função parar inserir os dados em uma lista de lista
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def transformaLista(lista):
  ListaDeItens = []
  cont = 1     #pulando o cabeçalho
  while cont< len(lista):
    ListaDeItens.append(transformaDados(lista[cont]))
    cont = cont + 1
  return ListaDeItens





#------------------------------------------------------------------ PARTE 2 - ENTRADA E VALIDAÇÃO DOS DADOS ---------------------------------------------------------

import matplotlib.pyplot as plt

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para entrada e validação dos dados informados pelo usuário (Intervalo Inicial)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def intervaloInicial():
  while True:
    mesInicial = int(input("Informe o mês inicial (de 1 a 12) que você deseja consultar:" ))
    if mesInicial < 1 or mesInicial > 12:
      print(">>Mês informado é inválido. Digite novamente!")
    else: break
  while True:
    anoInicial = int(input("Informe o ano inicial, entre 1961 e 2016, que você deseja consultar:" ))
    if anoInicial < 1961 or anoInicial > 2016:
      print(">>Ano informado fora do intervalo válido. Digite novamente!")
    else: break
  return mesInicial, anoInicial



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para entrada e validação dos dados informados pelo usuário (Intervalo Final)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def intervaloFinal():
  while True:
    mesFinal = int(input("Informe o mês final (de 1 a 12) que você deseja consultar:" ))
    if mesFinal < 1 or mesFinal > 12:
      print(">>Mês informado é inválido. Digite novamente!")
    else: break
  while True:
    anoFinal = int(input("Informe o ano final, entre 1961 e 2016, que você deseja consultar:" ))
    if anoFinal < 1961 or anoFinal > 2016:
      print(">>Ano informado fora do intervalo válido. Digite novamente!")
    else: break
  return mesFinal, anoFinal



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para filtragem dos dados dentro do intervalo informado pelo usuário
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def consultaDados(lista,mesInicial, anoInicial, mesFinal, anoFinal):#criação de uma função que consulta os dados na lista de acordo com as entradas do usuário
  dadosFiltrados = [] #criação de uma lista para armazenar os dados filtrados na consulta
  for item in lista:
    dia, mes, ano = item[0], item[1],item[2]
    if (ano > anoInicial or (ano == anoInicial and mes >= mesInicial)) and (ano < anoFinal or (ano == anoFinal and mes <= mesFinal)): #verifica se o mês e ano da lista estão dentro do intervalo informado pelo usuário
      dadosFiltrados.append(item)
  return dadosFiltrados



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para exibição de todos dados dentro do intervalo informado pelo usuário  - OPÇÃO 1
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def todosOsDados(dadosFiltrados, opcao):
  for item in dadosFiltrados:
    if opcao == 1:
      print(f"Data: {item[0]}/{item[1]}/{item[2]}\t\tPrecipitação: {item[3]}mm/m²\t\tTemp. Máxima: {item[4]}ºC\t\tTemp. Mínima: {item[5]}ºC\t\t Umidade: {item[6]}%\t\tVento: {item[7]}m/s")



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para exibição dos dados de precipitação dentro do intervalo informado pelo usuário - OPÇÃO 2
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def dadosPrecipitacao(dadosFiltrados, opcao):
  for item in dadosFiltrados:
    if opcao == 2:
      print(f"Data: {item[0]}/{item[1]}/{item[2]}\t\tPrecipitação: {item[3]}mm/m²")



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para exibição dos dados de temperatura dentro intervalo informado pelo usuário  - OPÇÃO 3
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def dadosTemperatura(dadosFiltrados, opcao):
  for item in dadosFiltrados:
    if opcao == 3:
      print(f"Data: {item[0]}/{item[1]}/{item[2]}\t\tTemp. Máxima: {item[4]}ºC\t\tTemp. Mínima: {item[5]}ºC")



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para exibição dos dados de Umidade e Vento informado pelo usuário - OPÇÃO 4
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def dadosUmidadeVento(dadosFiltrados, opcao):
  for item in dadosFiltrados:
    if opcao == 4:
      print(f"Data {item[0]}/{item[1]}/{item[2]}\t\tUmidade: {item[6]}%\t\tVento: {item[7]}m/s")


#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para transformar o número do mês em extenso para as impressões no console
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def nomeMesExt(numMes):
  if numMes == 1: nomeMes = "Janeiro"
  if numMes == 2: nomeMes = "Fevereiro"
  if numMes == 3: nomeMes = "Março"
  if numMes == 4: nomeMes = "Abril"
  if numMes == 5: nomeMes = "Maio"
  if numMes == 6: nomeMes = "Junho"
  if numMes == 7: nomeMes = "Julho"
  if numMes == 8: nomeMes = "Agosto"
  if numMes == 9: nomeMes = "Setembro"
  if numMes == 10: nomeMes = "Outubro"
  if numMes == 11: nomeMes = "Novembro"
  if numMes == 12: nomeMes = "Dezembro"
  return nomeMes

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Função para somar as precipitações por mês/ano e armazená-las em um dicionário - ITEM B do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def somaPrecipitacao(lista):
  precip = {} #criação de um dicionário para armazenar as precipitações
  for item in lista:
    mes = item[1]
    ano = item[2]
    precipitacao = item[3]
    nomeMes = nomeMesExt(mes)
    chave = nomeMes + str(ano) # determina que a chave será o mês por extenso mais o ano
    if chave in precip: #verifica se a chave (mesAno) já consta no dicionário, se sim, ele acumula a precipitação. Caso contrário, ele apenas inclui a precipitação no dicionário.
      precip[chave] += precipitacao
    else:
      precip[chave] = precipitacao
  return precip

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Mês mais chuvoso - ITEM B do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def mesMaisChuvoso(dicionario):
  maiorPrecip = 0 # criação da variável inicial para armazenar o valor da maior precipitação
  mesAno = "" # criação da variável inicial para armazenar o mês e ano da maior precipitação
  for chave, valor in dicionario.items(): # percorre o dicionário e verifica qual a maior precipitação, mantendo a chave (mesAno) junto
    if valor > maiorPrecip:
      maiorPrecip = valor
      mesAno = chave
  return mesAno, maiorPrecip

#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Média da temperatura mínima do mês informado pelo usuário - Nos últimos 11 anos (2006 a 2016) - ITEM C do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def medTemperaturaMin(lista, mesInformado):
  tempMin = {} #criação de um dicionário para armazenar as médias de temperatura mínima
  for num in range(2006,2017):
    somaTemp = 0 #criação da variável soma para armazenar a soma das temperaturas
    cont = 0
    for item in lista:
        mes = item[1]
        ano = item[2]
        tempMinim = item[5]
        # verifica se o ano está entre 2006 e 2016, e se mês corresponde ao mês informado pelo usuário
        if ano == num and mes == mesInformado:
            somaTemp += tempMinim  # soma as temperaturas mínimas
            cont += 1  # contabiliza o número de registros
    if cont > 0:  # verifica se tem registros
        mediaTempMin = somaTemp / cont  # calcula a média
        nomeMes = nomeMesExt(mesInformado)
        tempMin[nomeMes + str(num)] = mediaTempMin  # armazenamento no dicionário com chave (mês, ano) e média de temperatura correspondente
  return tempMin



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Gráfico com as médias de temperatura mínima do mês informado pelo usuário - Nos últimos 11 anos (2006 a 2016) - ITEM D do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def gerarGraficoTemp(dicionario): #função para percorrer o dicionário de média de temperatura mínima
  mesAno = [] #criação de uma lista para armazenar o dado de mês e ano consultados no dicionário
  tempMed = [] #criação de uma lista para armazenar o dado de média de temperatura consultados no dicionário
  for chave, valor in dicionario.items():
    mesAno.append(chave) #armazea os dados de mês e ano na lista
    tempMed.append(valor) #armazea os dados de média de temperatura na lista
  plt.bar(mesAno,tempMed, color="#C8A2C8")
  plt.title("Médias de Temperatura Mínima do mês informado entre os anos de 2006 e 2016")
  plt.xlabel("Mês/Ano")
  plt.xticks(rotation=45) #rotação do eixo x para que os dados não fiquem sobrepostos
  plt.ylabel("Temperatura Mínima ºC")
  plt.show()



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Média geral da temperatura mínima do mês informado pelo usuário - Nos últimos 11 anos (2006 e 2016) - ITEM E do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
def mediaGeralTempMin(dicionario): #criação da função que percorre o dicionário de média temperatura mínima criado e calcula a média geral de temperatura mínima para o mês informado
  cont = 0
  somaTempGeral =0 #criação da variável de soma para somar as médias temperaturas
  for chave,valor in dicionario.items():
    somaTempGeral += valor
    cont += 1
  if cont > 0:
    mediaTempGeral = somaTempGeral / cont #calcula a média
  return mediaTempGeral



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Seleção da opção dos dados para consulta pelo usuário
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
while True:
  print("\n\n\n\nCONSULTA DE DADOS METEOROLÓGICOS - ENTRE OS ANOS DE 1961 a 2016\n")
  print("Selecione os dados que você deseja consultar:")
  print()
  print("1 - Todos os dados")
  print("2 - Apenas os dados de precipitação")
  print("3 - Apenas os dados de temperatura")
  print("4 - Apenas os dados de umidade e vento")
  print("5 - Sair e consultar os dados de temperatura mínima entre os anos de 2006 e 2016")
  print()
  op = int(input())
  if op == 5:
   print("Fim desta consulta")
   break
  else:
    mesInicial, anoInicial = intervaloInicial()
    mesFinal, anoFinal = intervaloFinal()
    dadosBrutos = CargaDados("Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv") #chama a função para carregar os dados
    dados = transformaLista(dadosBrutos) #chama a função que transforma os dados conforme especificado
    dadosFiltrado = consultaDados(dados,mesInicial, anoInicial, mesFinal, anoFinal) #chama a função de consultada dos dados
    print()
    print(f"---------------- Dados referentes ao período de: {nomeMesExt(mesInicial)}/{anoInicial} a {nomeMesExt(mesFinal)}/{anoFinal} ----------------") #para exibir na tela o intervalo informado
    if op == 1:
      print()
      todosOsDados(dadosFiltrado, op)
    if op == 2:
      print()
      dadosPrecipitacao(dadosFiltrado, op)
      print()
      print("--------- Mês e Ano mais chuvoso entre os anos de 1961 a 2016 ---------")
      print()
      dadosBrutos = CargaDados("Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv") #chama a função para carregar os dados
      dados = transformaLista(dadosBrutos)#chama a função que transforma os dados conforme especificado
      dicionario = somaPrecipitacao(dados) #chama a função que armazena os dados de precipitação no dicionário
      mesAno, maiorPrecip = mesMaisChuvoso(dicionario) #chama a função que acessa o dicionário das precipitações e retorna a maior precipitação
      print(f"Mês/Ano com a maior precipitação: {mesAno}")
      print(f"Maior Precipitação: {maiorPrecip:.1f} mm/m²")
    if op == 3:
      print()
      dadosTemperatura(dadosFiltrado, op)
    if op == 4:
      print()
      dadosUmidadeVento(dadosFiltrado, op)



#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Seleção da opção dos dados para consulta pelo usuário - Nos últimos 11 anos (2006 a 2016) - PARA OS ITENS C , D e E do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
while True:
  print("\n\n\n\nCONSULTA DE DADOS METEOROLÓGICOS - TEMPERATURA MÍNIMA NOS ÚLTIMOS 11 ANOS (2006 a 2016)\n")
  print()
  mesInformado = int(input("Informe o mês de (1 a 12) que você deseja consultar:"))
  if mesInformado < 1 or mesInformado > 12:
    print(">>Mês informado é inválido. Digite novamente!")
  else: break


#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Chamada das funções para impressão dos dados no console - IENS C, D e E do projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------

# chama a função que calcula a média da temperatura mínima do mês informado pelo usuário, para cada ano entre 2006 e 2016. O resultado é um dicionário com as médias.
resultado = medTemperaturaMin(dados, mesInformado)

# calcula a média geral da temperatura mínima ao longo dos anos para o mês informado, conforme os dados do dicionário
mediaTempGeral = mediaGeralTempMin(resultado)

# chama a função para gerar o gráfico no dados do dicionário
gerarGraficoTemp(resultado)

#Uso da função que percorre o dicionário com as médias da temperatura mínima (entre os anos de 2006 e 2016) e calcula o valor da média geral de temperatura para o mês informado pelo usuário.
#Impressão da média geral de temperatura no console
print(f"Media Geral de Temperatura Mínima do mês {nomeMesExt(mesInformado)} nos últimos 11 anos (2006 a 2016) foi de: {mediaTempGeral:.1f}ºC")

#médias da temperatura mínima (entre os anos de 2006 e 2016) do mês informado pelo usuário
for chave, valor in resultado.items():
  mesAno = chave
  print(f"A média da temperatura mínima em {mesAno} foi de: {valor:.1f}ºC")


#---------------------------------------------------------------------------------------FIM--------------------------------------------------------------------------





#--------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Chamada das funções para conferências durante o projeto
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------

#consulta se o dicionário foi criado corretamente
#print(f"Média de Temperatura mensal:", medTemperaturaMin(dados,mesInformado))

#dadosBrutos = CargaDados("Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv")
#escreveLista(dadosBrutos)
#cabecalho = dadosBrutos[0].split(',')
#print(cabecalho)

#ListaDeItens = transformaLista(dadosBrutos)
#escreveLista(ListaDeItens)

#dadosFiltrados = transformaLista(dadosBrutos)
#escreveLista(dadosFiltrados)

```
